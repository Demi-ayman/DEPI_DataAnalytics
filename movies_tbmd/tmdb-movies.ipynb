```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Wrangling

```python
# load data
df = pd.read_csv('tmdb-movies.csv')
df.head(3)
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.drop(['id','imdb_id'],axis=1)
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# replace any zero to null to count no.of null values 
df[['revenue', 'budget']] = df[['revenue', 'budget']].replace(0, np.NAN)
df[['revenue_adj', 'budget_adj']] = df[['revenue_adj', 'budget_adj']].replace(0, np.NAN)
```

```python
df.isnull().sum()
```

```python
df = df.drop(['production_companies','tagline','keywords','homepage'],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Change the datatype of release date to dates datatype
df['release_date']=pd.to_datetime(df['release_date'])
df.info()
```

```python
#Calculate profit from budget and revenue
df['profit'] = df['revenue'] - df['budget']
df.head(1)
```

```python
#Function to count values in a column
def count_values(columns, column_value):
    count = 0
    for i in df[columns]:
        if column_value in i:
            count+=1
    return count
```

```python
#check for Adventure movies
print("there are",count_values('genres','Adventure'),"movies related to Adventure")
#check for Action movies
print("there are",count_values('genres','Action'),"movies related to Romantic ")
#check for Science Fiction movies
print("there are",count_values('genres','Science Fiction'),"movies related to Science Fiction ")
#check for Drama movies
print("there are",count_values('genres','Drama'),"movies related to Drama ")
#check for Thriller movies
print("there are",count_values('genres','Thriller'),"movies related to Thriller ")
#check for Crime movies
print("there are",count_values('genres','Crime'),"movies related to Crime ")
#check for Western movies
print("there are",count_values('genres','Western'),"movies related to Western ")
#check for Comedy movies
print("there are",count_values('genres','Comedy'),"movies related to Comedy ")
#check for Fantasy movies
print("there are",count_values('genres','Fantasy'),"movies related to Fantasy ")
#check for Romance movies
print("there are",count_values('genres','Romance'),"movies related to Romance ")
#check for Music movies
print("there are",count_values('genres','Music'),"movies related to Music ")
#check for War movies
print("there are",count_values('genres','War'),"movies related to War ")
#check for Mystery movies
print("there are",count_values('genres','Mystery'),"movies related to Mystery ")
```
